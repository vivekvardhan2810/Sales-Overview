```python
import matplotlib.pyplot as plt

# Sales data
days = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5', 'Day 6', 'Day 7']
week1_sales = [5000, 5900, 6500, 3500, 4000, 5300, 7900]
week2_sales = [4000, 3000, 5000, 5500, 3000, 4300, 5900]
week3_sales = [4000, 5800, 3500, 2500, 3000, 5300, 6000]

# Plotting the line chart
plt.plot(days, week1_sales, color='red', label='Week 1')
plt.plot(days, week2_sales, color='blue', label='Week 2')
plt.plot(days, week3_sales, color='brown', label='Week 3')

# Adding labels and title
plt.title('Sales Report')
plt.xlabel('Days')
plt.ylabel('Sales in Rs')

# Display legend
plt.legend()

# Display the line chart
plt.show()
```
